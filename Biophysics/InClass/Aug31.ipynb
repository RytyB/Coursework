```python
import numpy as np
import matplotlib.pyplot as plt
```

<h1> Demonstration of Simulation </h1>

```python
n = 10**4
obs = np.zeros(6)
for i in range(0,n):
    roll = np.random.randint(6)
    obs[roll] += 1
obs /= n
print(obs)
```

```python
# Standard plotting example
fig, ax = plt.subplots( figsize=(12,6) )
ax.plot(range(1,7), obs)
ax.grid()
ax.set_yticks( np.arange(0,1, .1) )
plt.show()
```

```python
# Standard plotting example
plt.bar(range(1,7), obs)
plt.show()
```
