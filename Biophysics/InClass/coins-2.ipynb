```python
import numpy as np
nflip=1000
flips=""
for i in range(1,nflip+1):
  if np.random.randint(0,2)==0:
    flips=flips+"H"
  else:
    flips=flips+"T\u0332"
  if i%50==0:
    flips=flips+"\n"

print(flips)
```

```python
import numpy as np
import matplotlib.pyplot as plt
ntry=10000
counts=np.zeros(30)
over=0
prob=.1
for attempt in range(ntry):
  hit=False
  n=0
  while(not hit):
    n+=1
    if np.random.random()<prob:
        hit=True
  if n<31:
    counts[n-1]+=1
  else:
    over+=1
plt.plot(counts)
plt.show()
```
