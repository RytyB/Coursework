```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def poisson(a, x):
    return a*np.exp(-a*x)
```

```python
# Demonstrate algorithm to sample a poisson distribution
nsims = int( 1e5 )
a = 4

r = []
for sim in range(0, nsims):
    r_iter = np.random.rand()
    r.append(r_iter)

r = np.array( r )
x = - np.log(r+1) / a
```

```python
# Binning to create a histogram
nbins = 50
counts, bins = np.histogram(x, bins=nbins)
counts = counts / np.sum( counts )

fig, ax = plt.subplots( figsize = (12,6) )
ax.bar(bins[:-1], counts, align='edge')
x_range = np.arange(0, 1, .1)
ax.plot(x_range, poisson(a, x_range), color = 'red')
plt.show()
```
