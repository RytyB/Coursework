```python
# what is the distribution of the number of flips required to get a head
import numpy as np
import matplotlib.pyplot as plt

#given probability = prob of seeing a head
# how many flips to see a head
prob=.1
max=40

ntry=50                    #number of trials
skip=0                      #how many results do we ignore?
counts=np.zeros(max)         #where we store the results

for attempt in range(ntry): #starting flips
  hit=False                 #we start with no successes
  n=0                       #we start with count = 0
  while(not hit):           #until we hit a target, we keep trying
    n+=1                    #update the number of tries
    if np.random.random()<prob:  #this is how we implement a non-uniform distribution
      hit=True            #we finally found the head!
  if n<max+1:                  #if the number of trials is within our storage limit
    counts[n-1]+=1          #increase the number of hits occurring at n
  else:                     #otherwise
    skip+=1                 #store the skipped count.
counts/=ntry
print("skipped "+str(skip)+" out of "+str(ntry))

predicted=np.array([prob*(1-prob)**k for k in range(0,max)])

xvalues=np.array(range(max))

plt.scatter(xvalues,counts)
plt.plot(xvalues,predicted)
plt.xlabel("flips")
plt.ylabel("P(flips)")
plt.show()
```

```python
# what is the distribution of the number of flips required to get a head
import numpy as np
import matplotlib.pyplot as plt

#given probability = prob of seeing a head
# how many flips to see a head
prob=.1
max=40



av=np.zeros(max)            #where we store the average of the simulations
var=np.zeros(max)           #where we store the sdtdev of the simulations

ntry=50                    #number of trials per simulation
nsim=100                    #number of simulations
skip=0                      #how many results did we ignore?

for simulation in range(0,nsim):
  counts=np.zeros(max)        #where we store the results per sim.  we reset every simulation to 0
  for attempt in range(ntry): #starting flips
    hit=False                 #we start with no successes
    n=0                       #we start with count = 0
    while(not hit):           #until we hit a target, we keep trying
      n+=1                    #update the number of tries
      if np.random.random()<prob:  #this is how we implement a non-uniform distribution
        hit=True            #we finally found the head!
    if n<max+1:                  #if the number of trials is within our storage limit
      counts[n-1]+=1          #increase the number of hits occurring at n
    else:                     #otherwise
      skip+=1                 #store the skipped count.
  counts/=ntry
  av+=counts
  var+=counts**2
av/=nsim
var/=nsim
var=(var-av**2)
stderr=np.sqrt(var/nsim)


print("skipped "+str(skip)+" out of "+str(ntry*nsim))

predicted=np.array([prob*(1-prob)**k for k in range(0,max)])

xvalues=np.array(range(max))

plt.scatter(xvalues,av,color='b')
plt.plot(xvalues,predicted,color='k')
plt.errorbar(xvalues, av, yerr = 2.98*stderr,fmt='', color='b')
plt.xlabel("flips")
plt.ylabel("P(flips)")
plt.show()
```
