```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Problem 1</h2>

```python
nsims = int( 1e6 )
r = []
for n in range(0,nsims):
    r.append( np.random.rand() )

r = np.array( r )
x = r ** 2

nbins = 500
counts, bins = np.histogram( x, bins=nbins, range=(0,1) )

fig, ax = plt.subplots( figsize = (12,6) )
ax.bar(bins[:-1], counts/nsims, width=(1/nbins), align='edge')
x_plot = np.arange(1/nbins, 1, 1/nbins)
ax.plot(x_plot, .001*x_plot**(-1/2), color = 'red', label = 'A * x**(-1/2)')
ax.set_ylim( 0,.01 )
ax.grid()
ax.set_xlabel( 'x = r^2' )
ax.set_ylabel( 'Simulated Probability' )
ax.legend(fontsize = 'xx-large')
plt.show()
```

<p> The distribution clearly has the form $ P(x) = A \frac1{\sqrt{x}} $, where the normalization contstant can be found by
normalizing the total probability to 1. </p>

<h2> Problem 2 </h2>

<p> I'm guessing that the average should happen at (b-a)*n </p>

```python
nsims = int( 1e5 )
nsteps = 10
a = 1
b = 2
finalPos = []


for i in range(0, nsims):
    position = 0
    for n in range(0, nsteps):
        right = np.random.randint(0,2)
        if right == 1:
            position += a
        else:
            position -= b
    finalPos.append( position )


fig, ax = plt.subplots( figsize=(12,6) )

max = np.max( [a,b] ) * n
nbins = 50
counts, bins = np.histogram( finalPos, bins=nbins, range=(- max, max) )

ax.bar( bins[:-1], counts, width=2*max/nbins, align='center' )
ax.grid()
plt.show()
```

<p> The average actually happens at $ pos = \frac{(a-b) \cdot n}{2} $. For every<b> two </b>steps, we should expect
to move an average distance of $ (a-b) $.
